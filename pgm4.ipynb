```python
import pandas as pd

data = pd.read_csv("training_data.csv")

h = ['0'] * (len(data.columns) - 1)

for _, row in data.iterrows():
    if row.iloc[-1] == "Yes":
        for i in range(len(h)):
            if h[i] == '0':
                h[i] = row.iloc[i]
            elif h[i] != row.iloc[i]:
                h[i] = '?'

print("Final Hypothesis:", h)
```
